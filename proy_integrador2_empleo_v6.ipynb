Para desarrollo de la funcion def calcular_anios_educacion(codigo): se utilizo como referencia el cuestionario de la encuesta de  Cuestionario EPM abril 2022 ¿Que grado de escolar más alto aprobó? Pregunta 6 image.png

```python
# =========================================
# Funciones auxiliares de recodificación
# =========================================

def clasificar_educacion_desde_texto(x):
    if pd.isna(x):
        return 'Inicial / Sin educación'

    if x in [
        'Ningún grado',
        'Prekínder o prejardín',
        'Kínder o jardín',
        'Enseñanza especial',
        'Menores de 4 años'
    ]:
        return 'Inicial / Sin educación'

    if x == 'Primaria':
        return 'Primaria'

    if x in ['Primer ciclo premedia', 'Segundo ciclo media']:
        return 'Media'

    if x == 'Vocacional':
        return 'Técnico'

    if x in ['Superior universitario', 'Superior no universitaria']:
        return 'Superior'

    if x in ['Especialidad', 'Maestría', 'Doctorado']:
        return 'Postgrado'

    return 'Otros'


def calcular_anios_educacion(codigo):
    try:
        codigo = int(codigo)

        if codigo <= 10:
            return 0

        # Primaria: 11 a 16
        if 11 <= codigo <= 16:
            return codigo - 10

        # Vocacional: 21 a 23
        if 21 <= codigo <= 23:
            return codigo - 14

        # Premedia: 31 a 33
        if 31 <= codigo <= 33:
            return codigo - 24

        # Media: 34 a 36
        if 34 <= codigo <= 36:
            return codigo - 24

        # Superior no universitaria: 41 a 42
        if 41 <= codigo <= 42:
            return codigo - 28

        # Superior universitaria: 51 a 56
        if 51 <= codigo <= 56:
            return codigo - 38

        # Postgrados
        if codigo == 61:
            return 19
        if 71 <= codigo <= 72:
            return 20
        if 81 <= codigo <= 84:
            return 22

        return 0

    except Exception:
        return 0


def recodificar_sector_eco_7(x):
    if pd.isna(x):
        return np.nan

    try:
        x = int(x)
    except Exception:
        return np.nan

    # 01. Agricultura, ganadería, silvicultura y pesca
    if x == 1:
        return 'Agricultura, ganadería, silvicultura y pesca'

    # 02. Minería; 04. electricidad y gas; 05. agua y residuos
    if x in [2, 4, 5]:
        return 'Minería y servicios públicos'

    # 03. Industrias manufactureras
    if x == 3:
        return 'Manufactura'

    # 06. Construcción
    if x == 6:
        return 'Construcción'

    # 07. Comercio; 08. transporte; 09. hoteles y restaurantes
    if x in [7, 8, 9]:
        return 'Comercio, transporte, alojamiento y comidas'

    # 10 a 14. Información, finanzas, inmobiliarias y servicios empresariales
    if x in [10, 11, 12, 13, 14]:
        return 'Servicios de información, financieros y empresariales'

    # 15 a 17. Administración pública, enseñanza y salud
    if x in [15, 16, 17]:
        return 'Administración pública y servicios sociales'

    # 18 a 21. Artes, otros servicios, hogares y organismos extraterritoriales
    if x in [18, 19, 20, 21]:
        return 'Otros servicios, hogares y actividades extraterritoriales'

    return np.nan


## Jornada Laboral semanal (horas laboradas/semana)
def clasificar_jornada(horas):
    if pd.isna(horas):
        return np.nan

    if horas < 15:
        return 'Tiempo parcial marginal'

    if horas < 35:
        return 'Tiempo parcial'

    if horas <= 48:
        return 'Jornada completa'

    return 'Horas excesivas'

    ## Recodificiación de estado civil
    df_analisis['estado_civil'] = pd.to_numeric(
    df_analisis['estado_civil'],
    errors='coerce'
    ).astype('Int64')

    df_analisis['estado_civil_rec'] = df_analisis['estado_civil'].map(dicc_estado_civil)
```

De acuerdo con la OIT:
- Menos de 15 horas/semana: trabajo a  **Tiempo parcial marginal** (marginal part-time).
- 15- 34 horas/semana: trabajo a **tiempo parcial** (part-time o short hours).
- 35 - 48 horas/semana: **jornada normal o completa**. La OIT considera 48 horas como el límite histórico de la jornada normal y promueve la reducción hacia 40 horas.
- Más de 48 horas/semana: **horas excesivas de trabajo** (long working hours o excessive working time).

```python
# =========================================
# Construcción de variables dentro de df_analisis
# =========================================

# Variables numéricas base
for col in [
    'Edad',
    'Sexo',
    'Jefe_de_hogar',
    'Asiste_a_la_escuela',
    'Grado_alcanzado',
    'Recibe_jubilacion_pension',
    'Salario_efectivo_empleado',
    'Salario_efectivo_independiente',
    'ingreso_hogar',
    'ingreso_per_capita',
    'dependencia'
]:
    if col in df_analisis.columns:
        df_analisis[col] = pd.to_numeric(df_analisis[col], errors='coerce')

# Salarios nulos como 0 para cálculos de ingreso laboral
df_analisis['Salario_efectivo_empleado'] = df_analisis['Salario_efectivo_empleado'].fillna(0)
df_analisis['Salario_efectivo_independiente'] = df_analisis['Salario_efectivo_independiente'].fillna(0)

# Sexo: conservar original, etiqueta para gráficos y dummy para modelos

df_analisis['Sexo_original'] = pd.to_numeric(
    df_analisis['Sexo'],
    errors='coerce'
)

df_analisis['Sexo_label'] = df_analisis['Sexo_original'].map({
    1: 'Masculino',
    2: 'Femenino'
})

df_analisis['sexo_bin'] = df_analisis['Sexo_original'].map({
    1: 1,
    2: 0
})

print(df_analisis['Sexo_original'].value_counts(dropna=False))
print(df_analisis['Sexo_label'].value_counts(dropna=False))
print(df_analisis['sexo_bin'].value_counts(dropna=False))

# Jefatura del hogar #Codificado sobre la encuesta del INEC
df_analisis['jefe_hogar_rec'] = np.select(
    [
        df_analisis['Jefe_de_hogar'] == 1,
        df_analisis['Jefe_de_hogar'] == 2,
        df_analisis['Jefe_de_hogar'] == 3,
        df_analisis['Jefe_de_hogar'] == 4,
        df_analisis['Jefe_de_hogar'].isin([5, 6])
    ],
    [1, 2, 3, 4, 5],
    default=np.nan
)

df_analisis['Jefe_de_hogar_label'] = df_analisis['Jefe_de_hogar'].map(dicc_jefe_h).fillna('Sin dato')

# Provincia: mantener código y etiqueta
df_analisis['provincia_cod'] = pd.to_numeric(df_analisis['provincia'], errors='coerce').astype('Int64')
df_analisis['provincia'] = df_analisis['provincia_cod'].astype(str).str.zfill(2).replace(dicc_provincia)

# Nivel educativo
df_analisis['grado_codigo'] = pd.to_numeric(df_analisis['Grado_alcanzado'], errors='coerce')
df_analisis['nivel_edu'] = df_analisis['grado_codigo'].map(dicc_nivel_edu).fillna('No aplica')
df_analisis['educacion_grupo'] = df_analisis['nivel_edu'].apply(clasificar_educacion_desde_texto)
df_analisis['anios_educacion'] = df_analisis['grado_codigo'].apply(calcular_anios_educacion)

# Edad al cuadrado
df_analisis['Edad_cuadrado'] = df_analisis['Edad'] ** 2

# Ingreso laboral total
df_analisis['ingreso_laboral'] = (
    df_analisis['Salario_efectivo_empleado'] +
    df_analisis['Salario_efectivo_independiente']
)

# Horas trabajadas y salario por hora
if 'Cuantas_horas_trabajó' in df_analisis.columns:
    df_analisis['horas_semana'] = pd.to_numeric(df_analisis['Cuantas_horas_trabajó'], errors='coerce')
else:
    df_analisis['horas_semana'] = np.nan

df_analisis['horas_mensuales'] = df_analisis['horas_semana'] * 4.33
df_analisis['salario_hora'] = df_analisis['ingreso_laboral'] / df_analisis['horas_mensuales']

# Logaritmos solo para valores positivos
df_analisis['ln_ingreso'] = np.nan
mask_ingreso = df_analisis['ingreso_laboral'] > 0
df_analisis.loc[mask_ingreso, 'ln_ingreso'] = np.log(df_analisis.loc[mask_ingreso, 'ingreso_laboral'])

df_analisis['ln_salario_hora'] = np.nan
mask_salario_hora = df_analisis['salario_hora'] > 0
df_analisis.loc[mask_salario_hora, 'ln_salario_hora'] = np.log(df_analisis.loc[mask_salario_hora, 'salario_hora'])

# Experiencia potencial
df_analisis['experiencia'] = (
    df_analisis['Edad'] -
    np.maximum(df_analisis['anios_educacion'] + 6, 18)
).clip(lower=0)

df_analisis['experiencia_cuadrado'] = df_analisis['experiencia'] ** 2

# Condición laboral
df_analisis['condicion_laboral'] = np.select(

    [

        df_analisis['Salario_efectivo_empleado'] > 0,
        df_analisis['Salario_efectivo_independiente'] > 0,
        df_analisis['Recibe_jubilacion_pension'] == 1,
        df_analisis['ocu_des'] == 'Desocupados',
        df_analisis['Asiste_a_la_escuela'] == 1,
        (
            (df_analisis['ocu_des'] == 'Ocupados') &
            (df_analisis['Salario_efectivo_empleado'] == 0) &
            (df_analisis['Salario_efectivo_independiente'] == 0)
        ),

        (
            df_analisis['ocu_des'].isna()
        )

    ],

    [

        'Empleado',
        'Independiente',
        'Jubilado / pensionado',
        'Desocupado',
        'Estudiante',
        'Ocupado sin ingreso declarado',
        'Inactivo económicamente'

    ],

    default='No declaró codición laboral'


)
df_analisis['tipo_ocupacion'] = np.select(
    [
        df_analisis['Salario_efectivo_empleado'] > 0,
        df_analisis['Salario_efectivo_independiente'] > 0
    ],
    [
        'Empleado',
        'Independiente'
    ],
    default='Sin ingreso laboral'
)


# Área geográfica
if 'Área_geográfica' in df_analisis.columns:
    df_analisis['area_geografica'] = df_analisis['Área_geográfica']
elif 'Area_geografica' in df_analisis.columns:
    df_analisis['area_geografica'] = df_analisis['Area_geografica']
else:
    df_analisis['area_geografica'] = np.nan

# Seguridad social
if 'Seguro_Social' in df_analisis.columns:
    seguro_social = pd.to_numeric(df_analisis['Seguro_Social'], errors='coerce')
    df_analisis['tiene_CSS'] = np.select(
        [
            seguro_social == 6,
            seguro_social.notna()
        ],
        [0, 1],
        default=np.nan
    )
else:
    df_analisis['tiene_CSS'] = np.nan

cols_seguridad_social = ['p72a', 'p72b', 'p72g1', 'p72g2', 'p72g3', 'p72g4', 'p72g5', 'p72g6']
cols_seguridad_social = [c for c in cols_seguridad_social if c in df_analisis.columns]

if cols_seguridad_social:
    df_analisis['seguridad_social'] = df_analisis[cols_seguridad_social].apply(
        pd.to_numeric,
        errors='coerce'
    ).fillna(0).sum(axis=1)
else:
    df_analisis['seguridad_social'] = np.nan

# Ingreso no laboral
cols_ingreso_no_laboral = [
    'p72c1', 'p72c2', 'p72c3', 'p72c4', 'p72c5', 'p72c7', 'p72c8',
    'p72f1', 'p72f2', 'p72f3', 'p72f4'
]
cols_ingreso_no_laboral = [c for c in cols_ingreso_no_laboral if c in df_analisis.columns]

if cols_ingreso_no_laboral:
    df_analisis['ingreso_no_laboral'] = df_analisis[cols_ingreso_no_laboral].apply(
        pd.to_numeric,
        errors='coerce'
    ).fillna(0).sum(axis=1)
else:
    df_analisis['ingreso_no_laboral'] = np.nan

# Tipo de contrato
if 'Tipo_de_contrato' in df_analisis.columns:
    tipo_contrato = pd.to_numeric(df_analisis['Tipo_de_contrato'], errors='coerce')
    df_analisis['tipo_contrato_eco'] = np.select(
        [
            tipo_contrato.isin([1, 4]),
            tipo_contrato.isin([2, 3]),
            tipo_contrato == 5
        ],
        [
            'Formal',
            'Temporal',
            'Informal'
        ],
        default='No especificado'
    )
else:
    df_analisis['tipo_contrato_eco'] = 'No especificado'

# Tamaño de empresa
if 'p29' in df_analisis.columns:
    p29_clean = pd.to_numeric(df_analisis['p29'], errors='coerce')
    df_analisis['tamano_empresa'] = p29_clean.map({
        1: 'Microempresa',
        2: 'Microempresa',
        3: 'Pequeña',
        4: 'Pequeña',
        5: 'Mediana_Grande'
    }).fillna('Sin respuesta')

    df_analisis['tamano_empresa_rec'] = p29_clean.map({
        1: 1,
        2: 1,
        3: 2,
        4: 2,
        5: 3
    }).fillna(0).astype(int)
else:
    df_analisis['tamano_empresa'] = 'Sin respuesta'
    df_analisis['tamano_empresa_rec'] = 0

# Sector económico
if 'Actividad_económica' in df_analisis.columns:
    df_analisis['actividad_economica_cod'] = pd.to_numeric(
        df_analisis['Actividad_económica'],
        errors='coerce'
    )
    df_analisis['sector_eco_7'] = df_analisis['actividad_economica_cod'].apply(recodificar_sector_eco_7)
else:
    df_analisis['sector_eco_7'] = np.nan

# Jornada laboral
df_analisis['jornada_dummy'] = np.where(
    df_analisis['horas_semana'] >= 40,
    1,
    np.where(df_analisis['horas_semana'].notna(), 0, np.nan)
)

df_analisis['tipo_jornada'] = np.select(
    [
        df_analisis['horas_semana'] >= 40,
        df_analisis['horas_semana'] < 40
    ],
    [
        'Jornada completa',
        'Jornada parcial'
    ],
    default='No especificado'
)

df_analisis['jornada_categoria'] = df_analisis['horas_semana'].apply(clasificar_jornada)

# Estado civil

if 'Estado_civil' in df_analisis.columns:

    estado_civil = pd.to_numeric(
        df_analisis['Estado_civil'],
        errors='coerce'
    ).astype('Int64')

    df_analisis['estado_civil_cod'] = estado_civil

    df_analisis['estado_civil_rec'] = np.select(
        [
            estado_civil.isin([1, 4]),
            estado_civil.isin([2, 3, 6]),
            estado_civil == 7,
            estado_civil == 5,
            estado_civil == 8
        ],
        [
            'Unido o casado',
            'Separado o viudo',
            'Soltero',
            'Divorciado',
            'Menor de 15 años'
        ],
        default='Sin dato'
    )

else:
    df_analisis['estado_civil_cod'] = np.nan
    df_analisis['estado_civil_rec'] = 'Sin dato'

# Antigüedad laboral
if 'p34' in df_analisis.columns:
    df_analisis['antiguedad_meses'] = pd.to_numeric(df_analisis['p34'], errors='coerce')
    df_analisis['antiguedad_anios'] = df_analisis['antiguedad_meses'] / 12
else:
    df_analisis['antiguedad_anios'] = np.nan

# Limpieza de infinitos
df_analisis = df_analisis.replace([np.inf, -np.inf], np.nan)

print("Variables principales creadas dentro de df_analisis.")
df_analisis.head(10)
```

```python
# =========================================
# V de Cramér entre variables categóricas
# =========================================

from scipy.stats import chi2_contingency
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Seleccionar variables categóricas de interés
variables_categoricas = [
    'Sexo',
    'educacion_grupo',
    'condicion_laboral',
    'provincia',
    'tipo_ocupacion',
    'area_geografica',
    'tipo_contrato_eco',
    'tamano_empresa',
    'tipo_jornada',
    'jornada_categoria'
]

# Mantener solo las que existan en df_analisis
variables_categoricas = [c for c in variables_categoricas if c in df_analisis.columns]

print("Variables categóricas usadas para V de Cramér:")
print(variables_categoricas)

# 2. Función para calcular V de Cramér corregida
def cramers_v(x, y):
    tabla = pd.crosstab(x, y)

    # Si la tabla está vacía o tiene una sola categoría en alguna dimensión, devolver NaN
    if tabla.shape[0] < 2 or tabla.shape[1] < 2:
        return np.nan

    chi2 = chi2_contingency(tabla)[0]
    n = tabla.sum().sum()

    if n == 0:
        return np.nan

    phi2 = chi2 / n
    r, k = tabla.shape

    # Corrección de sesgo
    phi2_corr = max(0, phi2 - ((k - 1) * (r - 1)) / (n - 1))
    r_corr = r - ((r - 1) ** 2) / (n - 1)
    k_corr = k - ((k - 1) ** 2) / (n - 1)

    denom = min((k_corr - 1), (r_corr - 1))

    if denom <= 0:
        return np.nan

    return np.sqrt(phi2_corr / denom)

# 3. Crear matriz de asociación categórica
matriz_cramer = pd.DataFrame(
    np.zeros((len(variables_categoricas), len(variables_categoricas))),
    index=variables_categoricas,
    columns=variables_categoricas
)

# 4. Calcular V de Cramér para cada par de variables
for var1 in variables_categoricas:
    for var2 in variables_categoricas:
        datos_temp = df_analisis[[var1, var2]].dropna()

        if var1 == var2:
            matriz_cramer.loc[var1, var2] = 1.0
        else:
            matriz_cramer.loc[var1, var2] = cramers_v(datos_temp[var1], datos_temp[var2])

# 5. Mostrar matriz numérica
print("\nMatriz de V de Cramér:")
display(matriz_cramer.round(3))

# 6. Heatmap
plt.figure(figsize=(12, 9))
sns.heatmap(
    matriz_cramer.astype(float),
    annot=True,
    fmt=".2f",
    cmap="YlOrRd",
    vmin=0,
    vmax=1,
    square=True,
    linewidths=0.5
)

plt.title("Matriz de asociación entre variables categóricas (V de Cramér)")
plt.tight_layout()
plt.show()
```

# Modelo Mincer

El modelo se estima usando `df_analisis`.  
Para la regresión se filtran únicamente observaciones válidas: ocupados, ingreso laboral positivo, horas trabajadas positivas y variables completas.

```python
# =========================================
# Crear variable participa
# =========================================

df_analisis['participa'] = np.where(
    df_analisis['ocu_des'].astype(str).str.strip().str.lower().str.contains('ocupado', na=False),
    1,
    0
)

print(df_analisis['ocu_des'].value_counts(dropna=False))
print(df_analisis['participa'].value_counts(dropna=False))
```

```python
# =========================================
# MODELO MINCER CORREGIDO
# =========================================

import statsmodels.api as sm
import numpy as np
import pandas as pd

# 1. Asegurar variable participa
df_analisis['participa'] = np.where(
    df_analisis['ocu_des'].astype(str).str.strip() == 'Ocupados',
    1,
    0
)

# 2. Asegurar años de educación
df_analisis['anios_educacion'] = pd.to_numeric(
    df_analisis['anios_educacion'],
    errors='coerce'
)

# 3. Experiencia potencial clásica de Mincer
df_analisis['experiencia'] = (
    df_analisis['Edad'] -
    df_analisis['anios_educacion'] -
    6
)

df_analisis['experiencia'] = df_analisis['experiencia'].clip(lower=0)
df_analisis['experiencia_cuadrado'] = df_analisis['experiencia'] ** 2

# 4. Ingreso laboral
df_analisis['ingreso_laboral'] = (
    df_analisis['Salario_efectivo_empleado'].fillna(0) +
    df_analisis['Salario_efectivo_independiente'].fillna(0)
)

# 5. Horas y salario hora
df_analisis['horas_semana'] = pd.to_numeric(
    df_analisis['Cuantas_horas_trabajó'],
    errors='coerce'
)

df_analisis['horas_mensuales'] = df_analisis['horas_semana'] * 4.33

df_analisis['salario_hora'] = (
    df_analisis['ingreso_laboral'] /
    df_analisis['horas_mensuales']
)

# 6. Logaritmo del salario hora
df_analisis['ln_salario_hora'] = np.where(
    df_analisis['salario_hora'] > 0,
    np.log(df_analisis['salario_hora']),
    np.nan
)

# 7. Muestra válida para Mincer
vars_mincer = [
    'ln_salario_hora',
    'anios_educacion',
    'experiencia',
    'experiencia_cuadrado',
    'sexo_bin'
]

mask_mincer = (
    (df_analisis['participa'] == 1) &
    (df_analisis['ingreso_laboral'] > 0) &
    (df_analisis['horas_semana'] > 0) &
    (df_analisis['horas_mensuales'] > 0)
)

datos_mincer = (
    df_analisis.loc[mask_mincer, vars_mincer]
    .replace([np.inf, -np.inf], np.nan)
    .dropna()
    .copy()
)

datos_mincer.to_csv('datos_mincer.csv', index=False)

print("Observaciones válidas para Mincer:", len(datos_mincer))
display(datos_mincer.describe())

# 8. Modelo Mincer básico
X_mincer = datos_mincer[
    [
        'anios_educacion',
        'experiencia',
        'experiencia_cuadrado',
        'sexo_bin'
    ]
]

X_mincer = sm.add_constant(X_mincer)
y_mincer = datos_mincer['ln_salario_hora']

modelo_mincer = sm.OLS(y_mincer, X_mincer).fit(cov_type='HC3')

print(modelo_mincer.summary())
```

# Modelo Heckman

Se mantiene `df_analisis` como dataset principal.  
La primera etapa estima la participación laboral mediante Probit y luego agrega la Razón Inversa de Mills (`IMR`) al mismo `df_analisis`.

```python
# =========================================
# Primera etapa Heckman: Probit de participación
# =========================================

import statsmodels.api as sm
from scipy.stats import norm

# -----------------------------
# 1. Variables
# -----------------------------
vars_seleccion = [
    'Edad',
    'Edad_cuadrado',
    'sexo_bin',
    'Asiste_a_la_escuela',
    'Recibe_jubilacion_pension',
    'niños',
    'adultos_mayores',
    'ingreso_hogar',
    'estado_civil_rec',
    'jefe_hogar_rec',
    'ingreso_no_laboral',
    'prov',
    'participa'
]

vars_seleccion = [c for c in vars_seleccion if c in df_analisis.columns]

datos_sel = df_analisis[vars_seleccion].copy()

print("Variables disponibles usadas:")
print(vars_seleccion)

# -----------------------------
# 2. Separar variable objetivo
# -----------------------------
y_col = 'participa'

if y_col not in datos_sel.columns:
    raise ValueError("La variable 'participa' no está en datos_sel.")

# -----------------------------
# 3. Crear dummies para variables categóricas
# -----------------------------
cat_cols = [
    'estado_civil_rec',
    'jefe_hogar_rec',
    'prov'
]

cat_cols = [c for c in cat_cols if c in datos_sel.columns]

datos_sel = pd.get_dummies(
    datos_sel,
    columns=cat_cols,
    drop_first=True,
    dtype=int
)

# -----------------------------
# 4. Convertir columnas numéricas
# -----------------------------
for col in datos_sel.columns:
    datos_sel[col] = pd.to_numeric(datos_sel[col], errors='coerce')

# -----------------------------
# 5. Revisar nulos antes de eliminar
# -----------------------------
print("\nNulos por variable antes de dropna:")
display(datos_sel.isna().sum().sort_values(ascending=False))

# -----------------------------
# 6. Eliminar nulos
# -----------------------------
datos_sel = datos_sel.dropna().copy()

print("\nFilas después de dropna:", len(datos_sel))

if len(datos_sel) == 0:
    raise ValueError("datos_sel quedó vacío después de dropna. Revisa variables con muchos nulos o texto convertido a NaN.")

# -----------------------------
# 7. Validar variable dependiente
# -----------------------------
print("\nDistribución de participa:")
print(datos_sel['participa'].value_counts(dropna=False))

if datos_sel['participa'].nunique() < 2:
    raise ValueError("La variable 'participa' necesita tener dos clases: 0 y 1.")

# -----------------------------
# 8. Definir X e y
# -----------------------------
X_sel = datos_sel.drop(columns=['participa'], errors='ignore')
X_sel = sm.add_constant(X_sel)

y_sel = datos_sel['participa']

# -----------------------------
# 9. Ajustar Probit
# -----------------------------
probit_model = sm.Probit(y_sel, X_sel)
probit_res = probit_model.fit()

print(probit_res.summary())
```

```python
# =========================================
# Razón Inversa de Mills
# =========================================

datos_sel['xb'] = probit_res.predict(X_sel, linear=True)
datos_sel['IMR'] = norm.pdf(datos_sel['xb']) / norm.cdf(datos_sel['xb'])
datos_sel['IMR'] = datos_sel['IMR'].replace([np.inf, -np.inf], np.nan)

# Guardar IMR dentro del dataset único
df_analisis['IMR'] = np.nan
df_analisis.loc[datos_sel.index, 'IMR'] = datos_sel['IMR']

df_analisis[['Edad', 'participa', 'IMR']].head(10)
```

```python
# =========================================
# Segunda etapa Heckman: ecuación salarial
# =========================================

vars_salario = [
    'ln_salario_hora',
    #'anios_educacion',
    'experiencia',
    #'experiencia_cuadrado',
    'sexo_bin',
    #'antiguedad_anios',
    'IMR',
    'tiene_CSS',
    'horas_mensuales',
    'educacion_grupo',
    'tipo_contrato_eco',
    'tamano_empresa',
    'area_geografica',
    'jornada_categoria',
    'estado_civil_rec',
    'sector_eco_7',
    'provincia',
    'participa'
]

vars_salario = [c for c in vars_salario if c in df_analisis.columns]

datos_wage = df_analisis.loc[df_analisis['participa'] == 1, vars_salario].copy()

cols_dummies = [
    c for c in [
        'educacion_grupo',
        'estado_civil_rec',
        'tipo_contrato_eco',
        'tamano_empresa',
        'jornada_categoria',
        'area_geografica',
        'sector_eco_7',
        'provincia'
    ] if c in datos_wage.columns
]

if cols_dummies:
    datos_wage = pd.get_dummies(
        datos_wage,
        columns=cols_dummies,
        drop_first=True,
        dtype=int
    )

for col in datos_wage.columns:
    datos_wage[col] = pd.to_numeric(datos_wage[col], errors='coerce')

datos_wage = datos_wage.replace([np.inf, -np.inf], np.nan).dropna().copy()

X_wage = datos_wage.drop(columns=['ln_salario_hora', 'participa'], errors='ignore')
X_wage = sm.add_constant(X_wage).astype(float)

y_wage = datos_wage['ln_salario_hora'].astype(float)

wage_model = sm.OLS(y_wage, X_wage).fit()

print(wage_model.summary())
```

#1. Importar librerías

```python
# ==========================================
# 1. LIBRERÍAS
# ==========================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

from scipy.cluster.hierarchy import linkage, dendrogram

sns.set_theme(style="whitegrid")
```

#2. Preparación de variables

```python
# ==========================================
# 2. PREPARACIÓN DE VARIABLES
# ==========================================

variables_cluster = [
    'Edad',
    'anios_educacion',
    'participa',
    'Asiste_a_la_escuela',
    'Recibe_jubilacion_pension',
    'condicion_laboral',
    'tiene_CSS',
    'tipo_contrato_eco',
    'tamano_empresa_rec',
    'area_geografica',
    'sector_eco_7',
    'ingreso_laboral',
    'horas_mensuales'
]

df_cluster = df_analisis[variables_cluster].copy()

df_cluster['horas_mensuales'] = df_cluster['horas_mensuales'].fillna(0)
df_cluster['sector_eco_7'] = df_cluster['sector_eco_7'].fillna('Sin actividad económica')

df_cluster['log_ingreso'] = np.log1p(df_cluster['ingreso_laboral'])
df_cluster['log_horas'] = np.log1p(df_cluster['horas_mensuales'])
```

#3. Matriz de segmentación

```python
# ==========================================
# 3. MATRIZ DE SEGMENTACIÓN
# ==========================================

variables_numericas = [
    'Edad',
    'anios_educacion',
    'log_ingreso',
    'log_horas'
]

variables_binarias = [
    'participa',
    'Asiste_a_la_escuela',
    'Recibe_jubilacion_pension',
    'tiene_CSS'
]

variables_categoricas = [
    'condicion_laboral',
    'tipo_contrato_eco',
    'tamano_empresa_rec',
    'area_geografica',
    'sector_eco_7'
]

X_num = df_cluster[variables_numericas + variables_binarias]

X_cat = pd.get_dummies(
    df_cluster[variables_categoricas],
    drop_first=False
)

X_final = pd.concat([X_num, X_cat], axis=1)

print("Matriz final:", X_final.shape)
print("Variables no numéricas:", X_final.select_dtypes(include=['object']).columns.tolist())
```

#4. Escalamiento y PCA

```python
# ==========================================
# 4. ESCALAMIENTO Y PCA
# ==========================================

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_final)

pca = PCA()
X_pca = pca.fit_transform(X_scaled)

varianza = pd.DataFrame({
    'Componente': range(1, len(pca.explained_variance_ratio_) + 1),
    'Varianza_Explicada': pca.explained_variance_ratio_,
    'Varianza_Acumulada': np.cumsum(pca.explained_variance_ratio_)
})

display(varianza.head(20))
```

```python
plt.figure(figsize=(10, 6))

plt.plot(
    varianza['Componente'],
    varianza['Varianza_Acumulada'],
    marker='o'
)

plt.axhline(y=0.80, linestyle='--', label='80%')
plt.axhline(y=0.90, linestyle='--', label='90%')

plt.xlabel('Número de componentes')
plt.ylabel('Varianza acumulada')
plt.title('Varianza explicada acumulada del PCA')
plt.legend()
plt.grid(alpha=0.4)
plt.tight_layout()
plt.show()
```

#5. Selección del número de clusters

```python
# ==========================================
# 5. SELECCIÓN DEL NÚMERO DE CLUSTERS
# ==========================================

pca15 = PCA(n_components=15, random_state=42)
X_pca15 = pca15.fit_transform(X_scaled)

resultados_k = []

for k in range(2, 11):
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=20
    )

    etiquetas = kmeans.fit_predict(X_pca15)

    resultados_k.append([
        k,
        silhouette_score(X_pca15, etiquetas),
        calinski_harabasz_score(X_pca15, etiquetas),
        davies_bouldin_score(X_pca15, etiquetas)
    ])

metricas_cluster = pd.DataFrame(
    resultados_k,
    columns=['k', 'Silhouette', 'Calinski_Harabasz', 'Davies_Bouldin']
)

display(metricas_cluster)
```

```python
plt.figure(figsize=(10, 6))

plt.plot(
    metricas_cluster['k'],
    metricas_cluster['Silhouette'],
    marker='o'
)

plt.xlabel('Número de clusters')
plt.ylabel('Silhouette')
plt.title('Evaluación del número de clusters')
plt.grid(alpha=0.4)
plt.tight_layout()
plt.show()
```

#9. Tabla final de perfiles

```python
# ==========================================
# 9. TABLA FINAL DE PERFILES LABORALES
# ==========================================

def moda_segura(x):
    return x.mode().iloc[0] if not x.mode().empty else np.nan

tabla_final = (
    df_cluster
    .groupby('perfil_laboral')
    .agg(
        N=('perfil_laboral', 'size'),
        Edad_promedio=('Edad', 'mean'),
        Años_educación=('anios_educacion', 'mean'),
        Ingreso_promedio=('ingreso_laboral', 'mean'),
        Ingreso_mediano=('ingreso_laboral', 'median'),
        Horas_mensuales=('horas_mensuales', 'mean'),
        Participación=('participa', 'mean'),
        Sexo_predominante=('Sexo_label', moda_segura),
        Área_predominante=('area_geografica', moda_segura),
        Sector_predominante=('sector_eco_7', moda_segura),
        Educación_predominante=('educacion_grupo', moda_segura)
    )
)

tabla_final['Porcentaje'] = (
    tabla_final['N'] / tabla_final['N'].sum() * 100
)

orden = [
    'Asalariados formales',
    'Trabajadores independientes',
    'Asalariados informales',
    'Trabajo no remunerado rural',
    'Inactivos y desocupados',
    'Estudiantes',
    'Jubilados y pensionados'
]

tabla_final = tabla_final.loc[orden].round(2)

display(tabla_final)
```

#10. Figura final: tipología laboral

```python
# ==========================================
# 10. FIGURA FINAL: TIPOLOGÍA LABORAL
# ==========================================

fig_tipologia = tabla_final.reset_index().copy()

macrosegmentos = {
    'Asalariados formales': 'Mercado laboral formal',
    'Trabajadores independientes': 'Mercado laboral vulnerable',
    'Asalariados informales': 'Mercado laboral vulnerable',
    'Trabajo no remunerado rural': 'Mercado laboral vulnerable',
    'Inactivos y desocupados': 'Fuera de la fuerza laboral',
    'Estudiantes': 'Fuera de la fuerza laboral',
    'Jubilados y pensionados': 'Fuera de la fuerza laboral'
}

fig_tipologia['Macrosegmento'] = fig_tipologia['perfil_laboral'].map(macrosegmentos)

plt.figure(figsize=(16, 8))

ax = sns.barplot(
    data=fig_tipologia,
    y='perfil_laboral',
    x='Porcentaje',
    hue='Macrosegmento',
    dodge=False
)

for container in ax.containers:
    ax.bar_label(
        container,
        fmt='%.2f%%',
        padding=4,
        fontsize=9,
        fontweight='bold'
    )

plt.title(
    'Tipología del mercado laboral panameño según perfiles de inserción económica',
    fontsize=16,
    fontweight='bold'
)

plt.xlabel('Porcentaje de la población analizada (%)')
plt.ylabel('Perfil laboral identificado')
plt.legend(title='Macrosegmento', loc='lower right')
plt.grid(axis='x', linestyle='--', alpha=0.35)
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# EXPORTAR DATASETS CLAVE DEL CLUSTER FINAL PARA POWER BI
# Pegar al final del bloque de clustering, después de df_pca
# ============================================================

import os
import zipfile
import pandas as pd

# ------------------------------------------------------------
# 1) CARPETA DE SALIDA
# ------------------------------------------------------------
EXPORT_DIR = "exports_cluster_final_powerbi"
os.makedirs(EXPORT_DIR, exist_ok=True)

# ------------------------------------------------------------
# 2) VALIDAR QUE LOS DATAFRAMES EXISTEN
# ------------------------------------------------------------
requeridos = {
    "df_cluster": "df_cluster.csv",
    "tabla_final": "tabla_final_perfiles.csv",
    "metricas_cluster": "metricas_cluster.csv",
    "df_pca": "df_pca_2d.csv"
}

faltantes = []
for nombre_df in requeridos.keys():
    if nombre_df not in globals():
        faltantes.append(nombre_df)

if faltantes:
    raise ValueError(
        f"Faltan estos objetos en memoria y por eso no se pueden exportar: {faltantes}. "
        "Asegúrate de ejecutar primero todo el bloque de cluster final."
    )

# ------------------------------------------------------------
# 3) PREPARAR LAS VERSIONES A EXPORTAR
# ------------------------------------------------------------
# tabla_final normalmente queda con perfil_laboral como índice;
# lo convertimos a columna para que Power BI lo lea mejor.
tabla_final_export = tabla_final.reset_index().copy()

# df_cluster ya contiene el detalle de registros + cluster + perfil_laboral
df_cluster_export = df_cluster.copy()

# metricas_cluster contiene la evaluación de k
metricas_cluster_export = metricas_cluster.copy()

# df_pca contiene PC1, PC2 y perfil_laboral
df_pca_export = df_pca.copy()

# ------------------------------------------------------------
# 4) EXPORTAR CSV CON ENCODING AMIGABLE PARA EXCEL/POWER BI
# ------------------------------------------------------------
archivos_exportados = []

exports = {
    "df_cluster.csv": df_cluster_export,
    "tabla_final_perfiles.csv": tabla_final_export,
    "metricas_cluster.csv": metricas_cluster_export,
    "df_pca_2d.csv": df_pca_export
}

for nombre_archivo, df_out in exports.items():
    ruta = os.path.join(EXPORT_DIR, nombre_archivo)
    df_out.to_csv(ruta, index=False, encoding="utf-8-sig")
    archivos_exportados.append(ruta)
    print(f"Exportado: {ruta} | filas={df_out.shape[0]} | columnas={df_out.shape[1]}")

# ------------------------------------------------------------
# 5) ZIP ÚNICO PARA DESCARGA MÁS FÁCIL
# ------------------------------------------------------------
zip_path = f"{EXPORT_DIR}.zip"
with zipfile.ZipFile(zip_path, "w", zipfile.ZIP_DEFLATED) as zipf:
    for ruta in archivos_exportados:
        zipf.write(ruta, arcname=os.path.basename(ruta))

print(f"\nZIP generado: {zip_path}")

# ------------------------------------------------------------
# 6) DESCARGA AUTOMÁTICA SI ESTÁS EN GOOGLE COLAB
# ------------------------------------------------------------
try:
    from google.colab import files
    files.download(zip_path)
except ImportError:
    print("\nNo estás en Google Colab.")
    print(f"Los CSV quedaron guardados en: {EXPORT_DIR}")
    print(f"El ZIP quedó guardado en: {zip_path}")
```
